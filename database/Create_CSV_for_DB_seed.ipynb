```python
import pandas as pd
import numpy as np
```

```python
gender = pd.read_csv('base_datasets/immigrants-emigrants-by-sex.csv')
gender
```

```python
nationality = pd.read_csv('base_datasets/immigrants-by-nationality.csv')
nationality
```

```python
age = pd.read_csv('base_datasets/immigrants-emigrants-by-age.csv')
age
```

### Create districts CSV

```python
##Use nationality original CSV as it is the one that includes all the district - neithbourhood combinations
districts_or = nationality.loc[:, 'District Code':'Neighborhood Name'].copy()
districts = districts_or.groupby(["District Code","District Name", "Neighborhood Code", "Neighborhood Name"]).count().reset_index()
districts
```

```python
districts.to_csv('CSV_for_DB_seed/districts_db_table.csv', sep ='\t', index=False, line_terminator = '\n')
```

### Create gender CSV

```python
gender_csv = gender[['Year', 'Neighborhood Code', 'Gender', 'Immigrants']].copy()
gender_csv
```

```python
gender_csv.to_csv('CSV_for_DB_seed/gender_db_table.csv', sep ='\t', index=False, line_terminator = '\n')
```

### Create age CSV

```python
age_csv = age[['Year', 'Neighborhood Code', 'Age', 'Immigrants']].copy()
age_csv
```

```python
age_csv.to_csv('CSV_for_DB_seed/age_db_table.csv', sep ='\t', index=False, line_terminator = '\n')
```

```python
nationality_csv = nationality[['Year', 'Neighborhood Code', 'Nationality', 'Number']].copy()
nationality_csv
```
