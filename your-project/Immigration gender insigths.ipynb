```python
import pandas as pd
import numpy as np
```

```python
gender = pd.read_csv('Percentage_gender_immigrant_population_year.csv')
gender.rename(columns={'percentage_immigrants_per_gender': 'percentage'}, inplace=True)
gender
```

```python
gender_district = pd.read_csv('Percentage_gender_immigrant_population_district_year.csv')
gender_district.rename(columns={'immigrants_percentage': 'percentage'}, inplace=True)
gender_district
```

### The average immigrant in terms of gender

The distribution of gender its almost 50-50 and it has followed the same trend in the last years:

```python
gender_distribution_years = gender
decimals = 2    
gender_distribution_years['percentage'] = gender_distribution_years['percentage'].apply(lambda x: round(x, decimals))
gender_years = gender_distribution_years.pivot(index="gender", columns="year", values="percentage")
gender_years
```

### Check gender distribution per district

Gender distribution per district. 
Most districts have a higher percentage of women than men.
Only 3 districts have a higher percentage of men (Sants-Montjuïc, Sant Martí, Ciutat Vella). The difference in Ciutat Vella is specially significant, men percentage being almost 10 points higher than women. 

```python
gender_district_2017 = gender_district[(gender_district['year'] == 2017)]
gender_district_2017 = gender_district_2017.loc[:, 'district_name':'percentage']
decimals = 2    
gender_district_2017['percentage'] = gender_district_2017['percentage'].apply(lambda x: round(x, decimals))
gender_district_2017
```

```python
gender_per_district_2017 = gender_district_2017.pivot(index="gender", columns="district_name", values="percentage")
gender_per_district_2017
```
