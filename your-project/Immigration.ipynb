# Immigrants per district per year in total numbers and as percentage

```python
def district_year(dis, year):
    filtered = immigrants[(immigrants['District Name'] == dis) & (immigrants['Year'] == year)]
    grouped = filtered.groupby('Nationality').sum()[['Number']].sort_values('Number', ascending=False)
    grouped['Year'] = year
    grouped['District_Name'] = dis
    

    return grouped
    
    
```

```python
years = sorted(immigrants['Year'].unique())
districts = sorted(immigrants['District Name'].unique())
```

```python
def percentages(dis,year):
    
    results = []
    for i in districts:
        for j in years:
            results.append((i, district_year(i, year)))

    district = [i[1] for i in results if i[0] == dis]
    district[0]['Percentage'] = district[0]['Number'].apply(lambda x: (x / district[0]['Number'].sum()) * 100)
    percent = district[0][district[0].Number != 0]
    return percent
```

# Immigrant nationality by district name

```python
def nat_year(nat, year):
    filtered = immigrants[(immigrants['Nationality'] == nat) & (immigrants['Year'] == year)]
    grouped = filtered.groupby('District Name').sum()[['Number']].sort_values('Number', ascending=False)
    grouped['Year'] = year
    grouped['Nationality'] = nat
    

    return grouped
```

```python
nat_year('Italy', 2017)
```

```python
nat_year('Pakistan',2017)
```
